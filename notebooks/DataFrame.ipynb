### Index操作

 ## .reset_index()で再度indexを割り振る

```python
df = pd.read_csv('../test-data/movies.csv')
df = df[df['original_language'] == 'ja'].reset_index(drop=True)
```

```python
df.reset_index()
```

```python
df = df.set_index('id')
df.head(3)
```

```python
## ilocではなく locを使う、理由：ilocはindexでしか使えないから
df.loc[192577]
```
